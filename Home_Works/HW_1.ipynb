## Задача 1

На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе, при помощи использования kwargs

````python
def total_friuts(**fruits):
    total = 0
    for amount in fruits.values():
        total += amount
    return total

total_friuts(banana=13, apple=8, grape=4, orange=18)
````

#### 3.2 
Найдите выручку компании в зависимости от месяца.
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, 
где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

````python
def calc_income_by_month(dates: list, incomes: list) -> dict:
    income_by_month = {}
    
    for i in range (len(dates)):
        month = dates[i].split('-')[1]
        if month in income_by_month.keys():
            income_by_month[month] += incomes[i]
        else:
            income_by_month[month] = incomes[i]
            
    return income_by_month
calc_income_by_month(dates, incomes)
        
        
        
#     df = pd.DataFrame()
#     df['Date'] = [datetime.datetime.strptime(i, '%Y-%m-%d') for i in data]
#     df['Sum'] = purchase
#     #df = df.groupby (df.Date.dt.month )['Sum']. sum ()
#     df['Month'] = pd.DatetimeIndex(df['Date']).month
#     df = df.groupby('Month').sum()
#     return df.to_dict('dict')
#     #df.to_dict('index')

# sum_month(dates, incomes)
````
